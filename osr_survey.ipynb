```python
import pandas as pd
```

```python
filename = 'OpenScienceRoom_survey_reponses.tsv'
responses = pd.read_csv(filename, sep='\t')
```

```python
questions = [
    'In the Open Science Room I would like to see',
    'I would like to see these topics in the OSR',
    'For educational content I go to the',
    'I go to the Open Science Room in order to',
]
question_columns = [[column for column in responses.columns if column.startswith(question)] for question in questions]
```

```python
#labels = ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree', 'N/A']
plot_axes = [
    responses[question_columns[index]]
        .apply(pd.Series.value_counts)
        .transpose()
        .plot
        .barh(stacked=True)
    for index, question in enumerate(question_columns)
];
```
